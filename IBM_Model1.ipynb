```python
en = open('/content/drive/My Drive/dev.en','r')
hi = open('/content/drive/My Drive/dev.hi','r')
sents_en=en.readlines()
sents_hi=hi.readlines()

 #initialize t(e|f) uniformly
 #dict[e,f] = value 
t = {}
for i in range(10):#doing it for 10 iterations talk about this 
  #  set count(e|f) to 0 for all e,f
  count = {}
  total = {}
  #  set total(f) to 0 for all f
  total_s = {}
  for i in range(len(sents_en)):
    for words_en in sents_en[i].split(' ')[:-1]:
      for words_hi in sents_hi[i].split(' ')[:-1]:
        if (words_en,words_hi) not in t:
          t[(words_en,words_hi)] = 0.2 # I have initialised to 0.2
        if words_en not in total_s:
          total_s[words_en] = 0
        total_s[words_en] += t[(words_en,words_hi)]
    for words_en in sents_en[i].split(' ')[:-1]:
      for words_hi in sents_hi[i].split(' ')[:-1]:
        if (words_en,words_hi) not in count:
          count[(words_en,words_hi)]=0
        if words_hi not in total:
          total[words_hi] = 0
        count[(words_en,words_hi)] += t[(words_en,words_hi)] / total_s[words_en]
        total[words_hi] += t[(words_en,words_hi)] / total_s[words_en]
  for key in t.keys():
    t[key] = count[key]/total[key[1]]
  i=1

words = []
for key in t.keys():
  if key[0] not in words:
    words.append(key[0])

results = {}
for word in words:
  max_val = -1
  max_key = ''
  for key in t.keys():
    if(key[0]==word) and max_val < t[key]:
      max_val = t[key]
      max_key = key
  results[max_key] = max_val
  # print(max_key)
  #iterate through all the key:value pairs with key[0]
# print(new_dict)
with open('ibmresults.pickle', 'wb') as handle:
    pickle.dump(results, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
